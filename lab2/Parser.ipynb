```python
from requests import get
#from currency_converter import CurrencyConverter
```

```python
def get_vac(salary, per_page = 100, pages = 20):
    vacancies = []
    
    for i in range(pages):
        url = 'https://api.hh.ru/vacancies'
        par = {'salary': salary, 'only_with_salary': 'true', 'per_page': per_page , 'page': i}
        r = get(url, params=par)
        e = r.json()
        vacancies.append(e)
        
    return vacancies
```

```python
vacs = get_vac(120000)
print(vacs[0])
```

```python
vacs = get_vac(15000)
print(vacs[0])
```

```python
def filter_vac(salary, less_than_salary=False):
    vacancies = get_vac(salary)
    jobs = {}
    key_skills = {}
    
    if less_than_salary:
        cmp = lambda x, y: x < y
    else:
        cmp = lambda x, y: x >= y
        
    for v in vacancies:
        items = v['items']
        
        for i in items:
            add_skills = False
            fr = i['salary']['from']
            to = i['salary']['to']
    
            if fr != None and to != None and cmp((fr + to) // 2, salary):
                jobs[i['name']] = [(fr + to) // 2, ' в среднем ', i['salary']['currency']]
                add_skills = True
            elif fr == None and cmp(to, salary):
                jobs[i['name']] = [to, ' до ', i['salary']['currency']]
                add_skills = True
            elif fr != None and cmp(fr, salary):
                jobs[i['name']] = [fr, ' от ', i['salary']['currency']]
                add_skills = True
            
            if add_skills:
                url = 'https://api.hh.ru/vacancies/' + str(i['id'])
                r = get(url)
                e = r.json()
                ks = e['key_skills']
                if(ks != None):
                    for skill in ks:
                        if skill['name'] not in key_skills: 
                            key_skills[skill['name']] = 1
                        else:
                            key_skills[skill['name']] += 1
    return jobs, key_skills
```

```python
good_jobs, good_key_skills = filter_vac(120000)
```

```python
for job in sorted(good_jobs.items(), key=lambda i: i[1][0]):
    print(job[0] + ', зарплата' + job[1][1] + str(job[1][0]))
```

```python
for skill in sorted(key_skills.items(), key=lambda i: i[1], reverse=True):
    print(skill[0] + ' - ' + str(skill[1]) + ' совпадений')
```

```python
bad_jobs, bad_key_skills = filter_vac(15000, True)
```

```python
for job in sorted(bad_jobs.items(), key=lambda i: i[1][0]):
    print(job[0] + ', зарплата' + job[1][1] + str(job[1][0]))
```

```python
for skill in sorted(bad_key_skills.items(), key=lambda i: i[1], reverse=True):
    print(skill[0] + ' - ' + str(skill[1]) + ' совпадений')
```
